# 1. Title: Home Math Intervention and Child’s Early Math Skills


# 2. **Background**



Mathematical knowledge has been proven to be an indicator of academic success, enrolment in scientific careers (Claessens & Engel, 2013) , and financial wellness (Storozuk & Maloney, 2023). According to the most recent National Assessment of Educational Progress Report (NAEP, 2025) , only 9% of fourth-grade students across the country mastered advanced math, 31% achieved proficient levels and 37% were placed in the basic math category followed by 24% students below the minimum standard. This is concerning because children underperforming in math from the beginning are likely to keep having difficulties in the subsequent years (Hooper et al., 2010; Judge & Watson, 2011) which might jeopardize their academic future.

The extent of math learning was previously attributed mostly to formal education (Bereiter & Engelmann, 1966). However, nowadays there is robust evidence in the developmental field, that the acquisition of these numerical competencies begins early on, and it is the reason why children exhibit differences in early math skills even before starting school (Jordan & Levine, 2009). Having strong numbers foundations will allow children to succeed in later math achievement (Melhuish et al., 2008).

Interventions based on home math environment were established due to the importance of parents and families as the first instance of children’s socialization and their role in supporting children’s interaction with numeracy-related content (LeFevre et al., 2009) such as quantities, number-words, and number visual representations, among others. Nonetheless, there has been conflicting evidence regarding whether or not children’s numeracy outcomes can be predicted by home math interventions and how relevant is this contribution (Daucourt et al., 2021; Zippert et al., 2020).

The current study aims to compare the efficacy of home math interventions to predict children’s early math capacities using everyday objects and activities that are incorporated easily into family routines. Accordingly, we hypothesized there would be a significant association between home interventions and child math skills (H1), particularly this effect would be higher for children who received  math intervention compared to control (H2). Lastly, we estimate a moderation effect of Count/cardinality domain in the relationship of intervention and math outcomes (H3).



# 3. **Variables**


*   Subject Id : (id) There are 76 participants
*   Age : (c_pre1_age) Child age in days at first pre-test since age it is a continous variable.
*   Sex : (c_female) In the dataset was coded as child is "1" female or "0" male because sex is a categorical variable.
*   Condition : (condition) Participants were randomly assigned to either math intervention (Math made for me) or control condition. In the data set was coded as "1" intervention "0" control because condition is a categorical variable.

  * Math Made 4 Me : 5-week video program designed for parents of preschool Children to promote home activities related to different math domains across their most frequent family routines . Each week, researchers sent a video introducing a different math domain, explaining the benefits of promoting the specific skill with children, and showing real examples of other parents implementing the program.

*   Math skills : We assessed children formal mathematical abilities using  The Preschool Early Numeracy Screener (PENS), a brief (24 items), and comprehensive measure of early numeracy for children aged 3 to 5. PENS effectively adjusts the difficulty according to the participant’s age by implementing a ceiling rule that stops the test when the child makes three consecutive mistakes (Purpura & Lonigan, 2015). There are 6 observations in the data set of this numeric variable per participant (N=76).

Math Domains measured by PENS:

  * Total number of correct counting and cardinality items (pre_total_countcard)
  * Total number of correct comparison items
  (pre_total_comp)
  * Total number of correct number identification items (pre_total_numid)
  * Total number of correct adding and subtracting items (pre_total_addsub)
  * Total number of correct patterning items(pre_total_patt)
  * Overall Math score before intervention: (pre_total_math)
  * Overall Math score after intervention: (post_total_math_24)












# 4. **Hypothesis**


**Pincipal Components (PCA)**

To better address which predictors should we include in the model, we will run a principal component analysis with the pre-test math domains and pre-executive functions measure. We anticipate that Component 1 overall math skills will explain most of the variance.

**Linear Regression Model**

There will be a condition effect of children assigned to the intervention "Math made 4 me". We hypothesize that after the intervention, children will increase their math skills (post-test), controlling for math skills pre-test. Thus, there will be an association of condition and post-math skills while controlling for pre-math skills.

Y (Post-math skills) = bO(intercept) + b1 (Condition) + b2 (Pre-math skills) + error.

**Moderation**

We estimate that previous experience with specific math domains would moderate children's math performance after the intervention. In particular, count and cardinality for preschooled children is been documented as one of the core math skills that scaffold the other domains. Therefore we hypothesize it would be a moderator.

Y (Post-math skills) = bO(intercept) + b1 (Condition) * Counting/Cardinality(pre_total_countcard) + b2 (Pre-math skills) + error.

- Principal Component Analysis (PCA)

```r
install.packages("pls")
library(pls)
library(gridExtra)
library(grid)

scaled_data <- scale(predictors)
pca_result <- prcomp(scaled_data, center = TRUE, scale. = TRUE)
loadings <- pca_result$rotation

options(repr.plot.width = 12, repr.plot.height = 6)

mathLoadings       <- loadings["pre_total_math", 1:6]
compLoadings       <- loadings["pre_total_comp", 1:6]
addsubLoadings     <- loadings["pre_total_addsub", 1:6]
countcardLoadings  <- loadings["pre_total_countcard", 1:6]
numidLoadings      <- loadings["pre_total_numid", 1:6]
pattLoadings       <- loadings["pre_total_patt", 1:6]


myLoadingsFrame <- round(t(data.frame(
  mathLoadings,
  compLoadings,
  addsubLoadings,
  countcardLoadings,
  numidLoadings,
  pattLoadings)), digits = 3)

colnames(myLoadingsFrame) <- paste0(1:6)
rownames(myLoadingsFrame) <- c(
  "Math Overall",
  "Comparison",
  "Addition/Subtraction",
  "Counting/Cardinality",
  "Number Identification",
  "Patterning")

myTheme <- ttheme_default(base_size = 18, base_colour = "black")

grid.table(myLoadingsFrame, theme = myTheme)
grid.draw(textGrob("PCA Loadings for Variables Table",
                   gp = gpar(fontsize = 20),
                   x = unit(0.62, "npc"), y = unit(0.8, "npc")))
grid.draw(textGrob("Components",
                   gp = gpar(fontsize = 18),
                   x = unit(0.30, "npc"), y = unit(0.66, "npc")))

```

```r
PCA_df <-clean_df[c("pre_total_math", "pre_total_comp", "pre_total_addsub", "pre_total_countcard", "pre_total_numid", "pre_total_patt" , "post_total_math_24")]

set.seed(2)
pcr_model <- pcr(
post_total_math_24 ~ .,
data = PCA_df,
scale = TRUE,
validation = "CV"
)


sD = c(1.784, 1.029, 0.854, 0.729, 0.703, 0.000)
propVar = c(0.5307, 0.1765, 0.1218, 0.0886, 0.0824, 0.0000)
cumVar = c(0.5307, 0.7072, 0.8290, 0.9176, 1.0000, 1.0000)

myPCASummaryFrame <- data.frame(
  `Standard Deviation` = sD,
  `Proportion of Variance` = round(propVar, 4),
  `Cumulative Proportion` = round(cumVar, 4)
)

rownames(myPCASummaryFrame) <- paste("Component", 1:6)


myTheme <- ttheme_default(base_size = 20, base_colour = "black")

grid.newpage()
grid.draw(textGrob("PCA Summary Table",
                   gp = gpar(fontsize = 22, fontface = "bold"),
                   x = 0.55, y = 0.75))
grid.table(myPCASummaryFrame, theme = myTheme)

validationplot(pcr_model, val.type = "MSEP")
```

```r
options(repr.plot.width = 8, repr.plot.height = 9)
library(ggplot2)

myValues <- c(0.5307, 0.7072, 0.8290, 0.9176, 1.0000, 1.0000)
myNumberofComp <- 1:6

myData <- data.frame(myNumberofComp, myValues)

ggplot(myData, aes(x = myNumberofComp, y = myValues)) +
  geom_bar(stat = "identity", fill = "darkorchid4", width = 0.8) +
  geom_text(aes(label = round(myValues, digits = 4)),
            vjust = 1.6, color = "white", size = 5) +
  theme_minimal(base_size = 14) +
  labs(
    x = "Components",
    y = "Cumulative Proportion of Variance Explained"
  ) +
  geom_hline(yintercept = 0.95, linetype = "dashed",color = "#E74C3C", linewidth = 1) +
  scale_x_continuous(breaks = 1:7) +
  ggtitle("Principal Components for 95% Variance Explained") +
  theme(
    plot.title = element_text(hjust = 0.5, size = 20),
    axis.text = element_text(size = 12),
    axis.title = element_text(size = 17)
  )
```

According to the results in the the PCA summary table and Cumulative Variance Explained Plot,  **Component 1 ** alone can explain the largest variance (about 53% of the variance) followed by Component 2 (17.6%). The variance portion increases until reaching the ceiling in component 5. Additionally, the PCA Loadings table suggests that the first principal component (Component 1) can ecompassed all the pre-math skills (Overall Math, Comparison, Addition/Subtraction, Counting/Cardinality, Number Identification, Patterning) and therefore it could be an underlying factor.

 Finally, the validation RMSEP indicated that for predicting math skills after the intervention (post-math skills), Component 1 is the most accurate explaining 50% of variance with the lowest prediction error.

Based on these results we will run a linear regression analysis with overall math skills as
the main predictor.

- Linear Regression Model for Prediction

```r
options(scipen=999)
model1 <- lm(post_total_math_24 ~ condition + pre_total_math , data = clean_df)
summary(model1)
```

There is a strong positive association among pre-math skills and post-math skills, indicating that previous experience with numbers will predict math skills in the future. For one unit in pre-math skill will result in 0.82 increase in post-math skills with a p-value = < .001. However, there is no difference between children who participated in the intervention and the ones assigned to control regarding their math skills as reflected for a p-value = > .05. Therefore, we cannot attribute the improvement in math skills to the intervention itself. The adjusted R2 is equivalent to 72% proving this model is better than the intercept.

```r

ggplot(clean_df, aes(x = pre_total_math, y = post_total_math_24 , color = condition)) +
  geom_point(alpha = 0.8) +
  geom_smooth(method = "lm", se = FALSE) +
  theme_classic() +
  labs(title = "Relationship between Pre and Post Math Skills by Condition",
       x = "Pre-overall Math Skills",
       y = "Post-overall Math Skills",
       color = "Condition") +
    scale_color_manual(
    values = c("1" = "purple4", "0" = "darkorchid1"),
    labels = c("1" = "Intervention", "0" = "Control")
  ) +
    theme(
    plot.title = element_text(size = 18, hjust = -0.3),
    axis.title = element_text(size = 15),
    axis.text = element_text(size = 15),
    legend.title = element_text(size = 15),
    legend.text = element_text(size = 15)
  )
```

- Generalized Linear Model : Moderation


```r
model2 <- lm(post_total_math_24 ~ condition * pre_total_countcard + pre_total_math , data = clean_df)
summary(model2)
```

```r
clean_df$predicted <- predict(model2, newdata = clean_df)

ggplot(clean_df, aes(x = pre_total_countcard, y = predicted, color = condition)) +
  geom_point(alpha = 0.5) +
  geom_smooth(method = "lm", se = FALSE) +
  theme_minimal() +
  labs(title = "Moderation Effect of Counting/Cardinality by Condition",
       x = "Pre Counting/Cardinality skills",
       y = "Post-overall Math",
       color = "Condition") +
    scale_color_manual(
    values = c("1" = "purple4", "0" = "darkorchid1"),
    labels = c("1" = "Intervention", "0" = "Control")
  ) +
    theme(
    plot.title = element_text(size = 18, hjust = -0.3),
    axis.title = element_text(size = 15),
    axis.text = element_text(size = 15),
    legend.title = element_text(size = 15),
    legend.text = element_text(size = 15)
  )
```

There is no interaction between counting/cardinality pre-math skills and condition while controlling for overall pre-math skills. We failed to reject the null hypothesis p = > .05. Once again, we found no intervention effect, just the association between pre-math skills and post-math skills from the previous linear model.


# 7. **Conclusions**

PCA analysis suggested that Component 1 would be sufficient to predict post-math skills (50%) and even thought, adding more components (up to 5) will increase this percentage at the same time would add more noise as well. In our linear model, we found an association between pre-overall math skills and post-math skills. However, this is not surprising because the same scale (PENS) was used before and after the condition (Intervention or control). We failed to reject the null hypothesis regarding the influence of condition on the post-math skill outcomes controlling for pre-math skills, and the reason behind it can be attributed to an ineffective intervention. Since this is a video based intervention, it is possible that parents could not understand or apply the different home environments exercises to their daily routine with their children. Another possibility is that the math skill screener (PENS) was inadequate and left no room for improvement to children in the study. Lastly, we found no interaction effects, between condition and Counting/cardinality math domain to predict post-math skills.
